```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
hvplot.extension('bokeh')
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
market_data_scaled = StandardScaler().fit_transform(df_market_data)
```

```python
# Create a DataFrame with the scaled data
df_market_data_scaled = pd.DataFrame(market_data_scaled, columns=['price_change_percentage_24h', 'price_change_percentage_7d', 'price_change_percentage_14d', 'price_change_percentage_30d', 'price_change_percentage_60d', 'price_change_percentage_200d', 'price_change_percentage_1y'])

# Copy the crypto names from the original data
df_market_data_scaled['coinid'] = df_market_data.index

# Set the coinid column as index
df_market_data_scaled = df_market_data_scaled.set_index('coinid')

# Display sample data
df_market_data_scaled.head(10)
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create an empty list to store the inertia values
inertia = []
# Create a list with the number of k-values from 1 through 11
k = list(range(1, 11))

# Create a for loop to compute the inertia with each possible value of k
for value in k:
  # 1. Create a KMeans model using the loop counter for the n_clusters
  kModel = KMeans(n_clusters=value, n_init='auto', random_state=1)
  # 2. Fit the model to the data using `df_market_data_scaled`
  kModel.fit(df_market_data_scaled)
  # 3. Append the model.inertia_ to the inertia list
  inertia.append(kModel.inertia_)
```

```python
# Create a DataFrame with the data to plot the Elbow curve
elbowDF = pd.DataFrame(
    {
        'K': k,
        'inertia': inertia
    }
)
```

```python
# Plot a line chart with all the inertia values computed with the different values of k to visually identify the optimal value for k.
hvplot.extension('bokeh')
elbowDFplot = elbowDF.hvplot.line(
                                  x='K',
                                  y='inertia',
                                  xticks = k
                                  )
elbowDFplot
```

#### Answer the following question:

**Question:** What is the best value for `k`?

**Answer:** There are two major elbows in the plot, so it appears that there are two possible values - either 3 or 5 clusters.

---

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialize the K-Means model using the best value for k
model = KMeans(n_clusters=3, n_init='auto')
```

```python
# Fit the K-Means model using the scaled data
model.fit(df_market_data_scaled)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
market_data_predictions = model.predict(df_market_data_scaled)

# Print the resulting array of cluster values.
market_data_predictions[:]
```

```python
# Create a copy of the DataFrame
df_market_data_scaled_predictions = df_market_data_scaled
```

```python
# Add a new column to the DataFrame with the predicted clusters
df_market_data_scaled_predictions['groups'] = market_data_predictions

# Display sample data
df_market_data_scaled_predictions.head(10)
```

```python
# Create a scatter plot using hvPlot by setting `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`.
# Color the graph points with the labels found using K-Means and add the crypto name in the `hover_cols` parameter to identify the cryptocurrency represented by each data point.

hvplot.extension('bokeh')
scaled_predictions_plot = df_market_data_scaled_predictions.hvplot.scatter(
    x='price_change_percentage_24h',
    y='price_change_percentage_7d',
    by='groups',
    hover_cols='coinid',
    title='Scaled Data Plot'
)
scaled_predictions_plot
```

---

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce to three principal components.
df_market_data_scaled_pca = pca.fit_transform(df_market_data_scaled_predictions)

# View the first five rows of the DataFrame.
df_market_data_scaled_pca[:5]
```

```python
# Retrieve the explained variance to determine how much information can be attributed to each principal component.
pca.explained_variance_ratio_
```

#### Answer the following question:

**Question:** What is the total explained variance of the three principal components?

**Answer:** The components add to around 89%.

```python
# Create a new DataFrame with the PCA data.
df_pca_data = pd.DataFrame(
    df_market_data_scaled_pca, columns=['PC1', 'PC2', 'PC3']
)

# Copy the crypto names from the original data
df_pca_data['coinid'] = df_market_data_scaled_predictions.index

# Set the coinid column as index
df_pca_data = df_pca_data.set_index('coinid')

# Display sample data
df_pca_data.head(10)
```

---

### Find the Best Value for k Using the PCA Data

```python
# Create an empty list to store the inertia values
inertia_2 = []
# Create a list with the number of k-values from 1 to 11
k_2 = list(range(1, 11))

# Create a for loop to compute the inertia with each possible value of k
for value in k_2:
  # 1. Create a KMeans model using the loop counter for the n_clusters
  kModel_2 = KMeans(n_clusters=value, n_init='auto', random_state=1)
  # 2. Fit the model to the data using `df_market_data_pca`
  kModel_2.fit(df_pca_data)
  # 3. Append the model.inertia_ to the inertia list
  inertia_2.append(kModel_2.inertia_)
```

```python
# Create a DataFrame with the data to plot the Elbow curve
elbowDF_2 = pd.DataFrame(
    {
        'K': k_2,
        'inertia': inertia_2
    }
)
```

```python
# Plot a line chart with all the inertia values computed with the different values of k to visually identify the optimal value for k.
hvplot.extension('bokeh')
elbowDF_2plot = elbowDF_2.hvplot.line(
    x='K',
    y='inertia',
    xticks=k_2,
    color='red'
)
elbowDF_2plot
```

#### Answer the following questions:

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** The best value for k is 4.


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** The best k value increased, whereas the inertia decreased across the board.

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Initialize the K-Means model using the best value for k
model_2 = KMeans(n_clusters=4, n_init='auto')
```

```python
# Fit the K-Means model using the PCA data
model_2.fit(df_pca_data)
```

```python
# Predict the clusters to group the cryptocurrencies using the PCA data
pca_data_predictions = model_2.predict(df_pca_data)
# Print the resulting array of cluster values.
pca_data_predictions[:]
```

```python
# Create a copy of the DataFrame with the PCA data
df_pca_data_predictions = df_pca_data

# Add a new column to the DataFrame with the predicted clusters
df_pca_data_predictions['groups'] = pca_data_predictions

# Display sample data
df_pca_data_predictions.head(10)
```

```python
# Create a scatter plot using hvPlot by setting `x="PC1"` and `y="PC2"`.
# Color the graph points with the labels found using K-Means and add the crypto name in the `hover_cols` parameter to identify the cryptocurrency represented by each data point.
hvplot.extension('bokeh')
pca_predictions_plot = df_pca_data_predictions.hvplot.scatter(
    x='PC1',
    y='PC2',
    by='groups',
    hover_cols='coinid',
    title='PCA Data Plot'
)
pca_predictions_plot
```
